```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from matplotlib.backends.backend_pdf import PdfPages
import numpy as np


bundesliga2 = pd.read_csv("bundesliga2.csv")
bundesliga = pd.read_csv("bundesliga.csv")
```

*** PART5 ***

SECTION POINT

```python
bundesliga2['winner'] = bundesliga2['POSITION'] == 1

seasons = bundesliga2['SEASON'].unique()

with PdfPages("Teams per season goal difference.pdf") as pdf:
    for season in seasons:
        season_data = bundesliga2[bundesliga2['SEASON'] == season]

        plt.figure(figsize=(10, 6))
        sns.barplot(data=season_data,x='DIFF',y='TEAM',hue='winner')

        plt.title(f"Season {season} Points")
        plt.xlabel("Points")
        plt.ylabel("Teams")
        plt.tight_layout()
        pdf.savefig()
        plt.close()
```




*** PART 6 ***

```python

sim_results = []
team_data = bundesliga2[bundesliga2['TEAM'] == "Bayern Munich"]
goals_per_season = team_data.groupby('SEASON')['GF'].sum()

for sim in range(10000):
    simulated_goals = np.random.choice(goals_per_season, size=10, replace=True)
    for year in range(10):
        sim_results.append({
            "Team": "Bayern Munich",
            "Season": year +2020,
            "Goals": simulated_goals[year]
        })

bayern_munich = pd.DataFrame(sim_results)


bayern_munich_grouped= bayern_munich.groupby(['Season'])['Goals'].mean().reset_index()

plt.figure(figsize=(10, 6))
sns.lineplot(data=bayern_munich,x='Season',y='Goals')
plt.title("average goals for Bayern Munich for Next 10 Seasons")
plt.xlabel("Season")
plt.ylabel("Average Goals")
plt.show()
```
